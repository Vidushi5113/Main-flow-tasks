```python
data.target.value_counts().plot(kind='bar')
plt.title("Herat Disease Values")
plt.xlabel("0 = No Disease, 1 = Disease")
plt.ylabel("Amount")
plt.show()
```

```python
data.target.value_counts().plot(kind='pie', figsize=(12,6))
plt.title('Target Distribution')
plt.legend(['Disease','No Disease'])
plt.show()
```

```python
data.sex.value_counts().plot(kind='pie', figsize=(12,6))
plt.title('Sex Distribution')
plt.legend(['Female','Male'])
plt.show()
```

```python
sns.countplot(x='target', hue='sex', data=data)
plt.title('heart disease frequency for sex')
plt.xlabel("0 = No Disease, 1 = Disease")
plt.show()
```

```python
data.cp.value_counts().plot(kind='bar')
plt.title('Chest Pain Type vs Count')
```

```python
pd.crosstab(data.sex,data.cp).plot(kind='bar')
plt.title('Type fo chest pain for sex')
plt.xlabel('Chest Pain Type')
plt.legend(['Female','Male'])
plt.show()
```
